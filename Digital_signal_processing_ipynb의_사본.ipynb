```python
def quantize(signal, levels):
  min_val = np.min(signal)
  max_val = np.max(signal)
  step = (max_val - min_val) / (levels - 1)
  quantized = np.round((signal - min_val) / step) * step + min_val
  return quantized

n = np.arange(0, 20)
x = np.sin(2 * np.pi * 0.1 * n)

xq = quantize(x, levels= 8)

plt.stem(n, xq, basefmt = " ", linefmt='r', markerfmt='ro')
plt.title("  (Discrete Signal,  )")
plt.xlabel("Sample Index (n)")
plt.ylabel("Quantized Amplitude")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = (9, 7)

sampFreq, sound = wavfile.read('sound1.wav')
sound.dtype, sampFreq
sound = sound / 2.0 ** 15
sound.shape

length_in_s = sound.shape[0] / sampFreq
print(length_in_s)

plt.subplot(2,1,1)
plt.plot(sound[:,0], 'r')
plt.xlabel("lefr channel samplr #")
plt.subplot(2,1,2)
plt.plot(sound[:,1], 'b')
plt.xlabel("right channel samplr #")
plt.tight_layout()
plt.show()

time = np.arange(sound.shape[0]) / sound.shape[0] * length_in_s

plt.snbplot(2,1,1)
plt.plot(time, sound[:,0], 'r')
plt.xlabel("Time, (s) [;efr channel]")
plt.ylabel("signal, relative units")
plt.subplot(2,1,2)
plt.plot(time, sound[:,1], 'b')
plt.xlabel("Time, (s) [;efr channel]")
plt.ylabel("signal, relative units")
plt.tight_layout()
plt.show()

signal = sound[:,0]

plt.plot(time[6000:7000], signal[6000:7000])
plt.xlabel("Time, (s)")
plt.ylabel("signal, relative units")
plt.show()

fft_spectrum = np.fft.rfft(signal)
freq = np.fft.rfftfreq(signal.size,d=1./sampFreq)
fft_spectrum

array([3686898804 + 0.j, 30.28378521-20.33641395j, 12.91719742-33.17116609j, ..., 1.4172224 + 1.4045349j, 1.8525152 + 0.76925206j, 1.99081421+0.j])
fft_spectrum_abs = np.abs(fft_spectrum)

plt.plot(freq, fft_spectrum_abs)
plt.xlabel("Frequency, (Hz)")
plt.ylabel("Amplitude, nuits")
plt.show()

plt.plot(freq[:3000],fft_spectrum_abs[:3000])
plt.xlabel("Frequency, (Hz)")
plt.ylabel("Amplitude, nuits")
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('sea.jpg', 0)
f_transform = np.fft.fft2(image)
f_transform_shifted = np.fft.fftshift(f_transform)
magnitude_spectrum = np.log(np.abs(f_transform_shifted)+1)

plt.subplot(121), plt.imshow(image, cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title("Magnityde Spectrum"), plt.xticks([]), plt.yticks([])
plt.show()
```
